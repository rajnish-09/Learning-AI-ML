```python
import numpy as np
```

```python
# Aggregation operations
arr = np.array([1,2,3,4,5,6,7])
# To find sum of all the elements
print(np.sum(arr))    
# To find product of all elements
print(np.prod(arr))
# To find min value 
print(np.min(arr))
# To find max value
print(np.max(arr))
# Mean
print(np.mean(arr))
# Standard deviation
print(np.std(arr))
# Variance 
print(np.var(arr))
# Median 
print(np.median(arr))
# To find min value index
print(np.argmin(arr))
# To find max value index
print(np.argmax(arr))
```

```python
# Power
arr = np.array([1,2,3,4,5,6,7])
# To print squares
print(np.square(arr))
# To find square root
print(np.sqrt(arr))
# To find power of 
print(np.pow(arr, 3))
```

```python
# Rounding
# To find round up number
print(np.round(4.6))
# To find lower value
print(np.floor(4.6))
# To find higher value
print(np.ceil(4.1))
# To truncate decimal values
print(np.trunc(3.456))
```

```python
# Log and Exponential
arr = np.array([1,2,3,4, 4,2,5,6,7])
# To find log
print(np.log(arr))
print(np.log10(arr))
print(np.log2(arr))

# To find exponential
print(np.exp(arr))
```

```python
# Extra
arr = np.array([5,3,1,2,3,4, 4,2,5,2,2,3])
# To find unique values only
print(np.unique(arr))
# To sort elements
print(np.sort(arr)
# 
```
