```python
import pandas as pd
```

```python
df = pd.read_csv("raw_data.csv")

# groupby country and find mean of salary
df.groupby("country")["income"].mean()
```

```python
# groupby country and find minimum salary
df.groupby("country")["income"].min()
```

```python
# groupby country and find minimum salary
df.groupby("country")["income"].max()
```

```python
# multiple aggregate
df.groupby("country")["income"].agg(["mean", "min", "max"])
```

```python
# mulitple aggregate renaming the column name
df.groupby("country")['income'].agg(avg_salary="mean", min_salary="min")
```

```python
# multiple aggregate with different operation in different column
df.groupby("country")['income'].agg(
    avg_salary=("income","mean"),
    min_salary=("income", "min")
)
```
