```python
import pandas as pd
```

```python
df = pd.read_csv("globalAirQuality.csv")
# To print data whose aqi > 100
df [ df['aqi'] > 100]
```

```python
# To get data whose aqi > 100 and temp > 30
df [ (df['aqi'] > 100) & (df['temperature'] > 30)]
```

```python
# Chaining -> Get only required datas after filtering
df [ (df['aqi'] > 100) & (df['tempertaure'] > 30)] [['city', 'aqi', 'temperature']]
```
