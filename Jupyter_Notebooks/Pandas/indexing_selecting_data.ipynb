```python
import pandas as pd
```

```python
df = pd.read_csv("globalAirQuality.csv")
# Indexing
# df['country']
df[['city', 'aqi']]
```

```python
# Selecting data
# 1. Using label (loc)
#df.loc[0]        # returns first row data
df.loc[0:2]       # start_index, end_index (end_index is incluscive in loc)

# Selecting individual cell
df.loc[0, 'city']
```

```python
# 2. Using iloc
# df.iloc[0]
df.iloc[0:3]       # start_index, end_index (end_index is exclusive in iloc)

# Accessing individual cell
# df.iloc[0, 2]
df.iloc[0, 1:4]
```

```python
# Selecting single scalar value
# 1. Using at
df.at[0, 'city']
```

```python
# 2. Using iat
df.iat[0, 2]
```
