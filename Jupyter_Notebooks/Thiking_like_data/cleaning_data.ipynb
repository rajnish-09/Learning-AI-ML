```python
import json
```

```python
def load_data(filename):
    with open(filename, 'r') as f:
        data = json.load(f)
        return data
data = load_data("customer_data.json")
```

```python
def clean_data(data):
    text_to_num = {"one":1, "two":2, "three":3, "four":4, "five":5}
    cleaned_data = []
    unique_data = set()
    for user in data:
        # Cleaning score
        raw_score = str(user['score']).strip().lower()
        if raw_score in text_to_num:
            user['score'] = text_to_num[raw_score]
        else:
            user['score'] = raw_score

        # Handling missing values
        raw_year = user.get("years")
        if raw_year == None:
            user['years'] = None

        # Handling duplicate values
        if user['customer'] in unique_data:
            continue
        unique_data.add(user['customer'])
        cleaned_data.append(user)
    return cleaned_data
```

```python
clean_data(data)
```
