```python
import numpy as np
import time
```

```python
# Calclulating the speed between python list and numpy array
size = 10_000_000
# python list
py_list = list(range(size))
start = time.time()
sq_list = [i**2 for i in py_list]
end = time.time()
print(f'Python list time = {end-start}')

# numpy array
np_arr = np.array(py_list)
start = time.time()
sq_numpy = np_arr ** 2
end = time.time()
print(f'Numpy array time = {end-start}')
```

```python
# Memory
import sys
print(f'Python list size = {sys.getsizeof(py_list) * len(py_list)} bytes')
print(f'Numpy array size = {np_arr.nbytes} bytes')
```
