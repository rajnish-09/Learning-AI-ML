```python
import json
```

```python
def load_data(filename):
    with open(filename, 'r') as f:
        data = json.load(f)
    return data
data = load_data("store_data.json")
```

```python
# Clean & structure data
def clean_data(data):
    text_to_num = {"one":1, "two":2, "three":3, "four":4, "five":5}
    cleaned_data = []
    unique_data = set()
    for user in data:
        # Cleaning rating
        raw_rating = str(user["rating"]).strip().lower()
        if raw_rating in text_to_num:
            raw_rating = text_to_num[raw_rating]
        user['rating'] = raw_rating

        # Handling missing values
        raw_age = user.get("age")
        if raw_age == None:
            user['age'] = None

        # Handling duplicate values
        if user['name'] in unique_data:
            continue
        unique_data.add(user['name'])
        cleaned_data.append(user)
        
    return cleaned_data

    
```

```python
data = clean_data(data)
```

```python
# Getting meaningful insights from data
def get_insights(data):
    # Calculating average rating
    total_rating = 0
    for user in data:
        total_rating += float(user['rating'])
    print(f'Average rating: {total_rating/len(data)}')

    # Calculate % of user with poor rating
    poor_rating = 0
    for user in data:
        if float(user['rating']) < 3:
            poor_rating += 1
    print(f'% of user with poor rating: {poor_rating/len(data) * 100} %')
```

```python
get_insights(data)
```

```python
# Recommendating product based on rating
def get_recommendations(data):
    recommendations = []
    for user in data:
        current_recom = {}
        current_recom['name'] = user['name']
        if float(user['rating']) >= 4:
            current_recom['brand'] = 'Apple'
        else:
            current_recom['brand'] = 'Samsung'
        recommendations.append(current_recom)
    return recommendations
```

```python
get_recommendations(data)
```
