```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "order_id": [1,2,3,4,5,6,7,8],
    "category": ["Electronics","Clothing","Electronics","Food",
                 "Food","Clothing","Electronics","Food"],
    "amount": [500,100,300,50,80,150,400,60]
})
df
```

```python
# Group by category.
# Calculate:
# Total sales amount.
# Average order amount.
# Maximum order amount.
# Number of orders.
# Reset the index.

df2 = df.groupby("category").agg({
    "amount" : ["sum", "mean", "max", "count"]
})
df2
```

```python
df2.columns = [
    "Total Sales",
    "Average Order",
    "Maximum Order",
    "Number of Orders"
]
df2
```

```python
# Sort by total sales descending
df2.sort_values("Total Sales", ascending=False)
```

```python
# save to excel file
df2.to_excel("category_sales_summary.xlsx")
```
