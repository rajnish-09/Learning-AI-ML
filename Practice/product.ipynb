```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "product": ["Laptop", "Phone", "Laptop", "Tablet", "Phone", "Tablet"],
    "sales": [5, 10, 3, 4, 7, 6],
    "revenue": [5000, 7000, 3000, 2000, 4900, 3000]
})
df
```

```python
# group by product and calculate total sales and revenue and average revenue
df = df.groupby("product").agg(
    total_sales = ("sales", "sum"),
    total_revenue = ("revenue", "sum"),
    avg_revenue = ("revenue", "mean")
)
df
```

```python
# short by revenue (highest first)
df.sort_values("total_revenue", ascending=False)
```

```python
# save to csv file
df.to_csv("product_sales_report.csv")
```
