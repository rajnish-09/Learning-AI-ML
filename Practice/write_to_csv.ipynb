```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "employee": ["John", "Alice", "Bob", "Emma", "David", "Sophia"],
    "department": ["IT", "HR", "IT", "Finance", "HR", "Finance"],
    "salary": [60000, 50000, 70000, 80000, 55000, 75000]
})
df
```

```python
# group by department
# find total salary and average salary per department and number of employees per departnment
df.groupby("department").agg(
    total_salary = ("salary", "sum"),
    avg_salary = ("salary", "mean"),
    no_of_employees = ("employee", "count")
)
```
